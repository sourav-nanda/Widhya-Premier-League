```python
def cat_plot(df,columns,size,plot_type):
    length=len(columns)
    fig,ax=plt.subplots(ncols=length,
                        figsize=size)
    
    for i in range(length):
        df[columns[i]].value_counts().plot(kind=plot_type,
                                           ax=ax[i],
                                           colors=['violet',
                                                   'pink'],
                                           autopct='%2.f%%')
        ax[i].set_title(columns[i],
                        fontsize=20)
```

```python
cat_plot(mi_dc,['winner','toss_winner'],(16,10),'pie')
```

```python
def cat_plot(df,columns,size,plot_type):
    length=len(columns)
    fig,ax=plt.subplots(nrows=length,
                        figsize=size)
    
    for i in range(length):
        df[columns[i]].value_counts().plot(kind=plot_type,
                                           ax=ax[i],
                                           color='darkviolet',
                                           fontsize=13)
        ax[i].set_title(columns[i],
                        fontsize=20)
        
```

```python
cat_plot(matches,['winner','toss_winner'],(12,15),'barh')
```

### As the head on data suggests there is 50-50 win chance for both teams and Dc is often a toss winner compared to Mi our first assumption should be that DC has higher chances of winning but if we check the overall wins then MI outshines DC by a heavy margin.

# Q2

```python
rabada_wickets=deliveries.query('bowler=="K Rabada"').groupby(['match_id']).player_dismissed.count()
rabada_wickets.describe()
```

```python
f,ax=plt.subplots(ncols=2,
                  figsize=(17,6))

sns.countplot(rabada_wickets,
              color='rebeccapurple',
              ax=ax[0])

sns.boxplot(rabada_wickets,
            color='plum',
            ax=ax[1])

f.suptitle('Kagiso Rabada wickets Frequency Distribution',fontsize=19)
```

### We can infer from the following distributions that he will take a total of 2 wickets

```python
mi_dcdeli[mi_dcdeli.batsman.str.contains('Pandya')].batsman.unique()
```

```python
pandya_runsoverall=deliveries.query('batsman=="HH Pandya"').groupby('match_id').batsman_runs.sum()
pandya_runsoverall.describe()
```

```python
pandya_runs=mi_dcdeli.query('batsman=="HH Pandya"').groupby('match_id').batsman_runs.sum()
pandya_runs.describe()
```

```python
f,ax=plt.subplots(ncols=2,
                  figsize=(17,6))

sns.distplot(pandya_runsoverall,
              color='mediumvioletred',
              ax=ax[0])
ax[0].set_title('Overall',fontsize=16)

sns.distplot(pandya_runs,
            color='palevioletred',
            ax=ax[1])
ax[1].set_title('Head on',fontsize=16)

f.suptitle('Hardik Pandya runs Frequency Distribution',fontsize=19)
```

### We can observe from the above distributions that Pandya will score between 21-30 runs
