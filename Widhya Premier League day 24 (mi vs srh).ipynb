```python
def cat_plot(df,columns,size,plot_type):
    length=len(columns)
    fig,ax=plt.subplots(ncols=length,figsize=size)
    for i in range(length):
        df[columns[i]].value_counts().plot(kind=plot_type,ax=ax[i],colors=['pink','violet'],autopct='%2.f%%')
        ax[i].set_title(columns[i])
```

```python
cat_plot(mi_srh,['winner','toss_winner'],(15,8),'pie')
```

```python
def cat_plot(df,columns,size,plot_type):
    length=len(columns)
    fig,ax=plt.subplots(nrows=length,figsize=size)
    for i in range(length):
        df[columns[i]].value_counts().plot(kind=plot_type,ax=ax[i],color='teal')
        ax[i].set_title(columns[i])
```

```python
cat_plot(matches,['winner','toss_winner'],(12,15),'barh')
```

# Q4

```python
balls_bowled=mi_srhdeli.groupby(['match_id',
                                 'bowler']).ball.count()

boundaries_scored=mi_srhdeli.query('batsman_runs>=4').groupby(['match_id',
                                                               'bowler']).ball.sum()
```

```python
boundary_leaker=1/(balls_bowled/(boundaries_scored+1))
boundary_leaker=boundary_leaker.groupby('bowler').mean()
```

```python
boundary_leaker.sort_values().plot.barh(figsize=(14,17),
                                        color='cornflowerblue')
plt.title('Boundary Leaker metric for bowlers of MI and SRH',fontsize=17)
```
